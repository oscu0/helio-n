```python
hpc_coords = all_coordinates_from_map(aiamap)

```

```python
mask = coordinate_is_on_solar_disk(hpc_coords)
palette = aiamap.cmap.copy()
palette.set_bad('black')
```

```python
scaled_map = sunpy.map.Map(aiamap.data, aiamap.meta, mask=~mask)

scaled_map.data
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection=scaled_map)
scaled_map.plot(axes=ax, cmap=palette)
scaled_map.draw_limb(axes=ax)
plt.show()
```
